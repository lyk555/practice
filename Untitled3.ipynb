```python
import networkx as nx
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
# Create a graph from the geospatial data
def create_graph_from_gdf(gdf):
    g = nx.Graph()
    for index, row in gdf.iterrows():
        g.add_node(index, feature=row.geometry.centroid.coords[0], label=row['label'])
    return g
```

```python
# Encode the 'CODICE' and 'COD_TOT' columns
le_1999 = LabelEncoder()
milano_1999['label'] = le_1999.fit_transform(milano_1999['CODICE'])

le_2021 = LabelEncoder()
milano_2021['label'] = le_2021.fit_transform(milano_2021['COD_TOT'])
```

```python
# Create graphs for both years
graph_1999 = create_graph_from_gdf(milano_1999)
graph_2021 = create_graph_from_gdf(milano_2021)
```

```python
# Function to add edges based on spatial proximity using KD-Tree
def add_edges_with_kdtree(graph, kdtree, threshold):
    for node in graph.nodes():
        coord = graph.nodes[node]['feature']
        neighbor_indices = kdtree.query_ball_point(coord, threshold)
        for neighbor in neighbor_indices:
            if neighbor != node:
                graph.add_edge(node, neighbor)
```

```python
from scipy.spatial import KDTree
```

```python
# Build KD-trees and add edges
kdtree_1999 = KDTree([graph_1999.nodes[node]['feature'] for node in graph_1999.nodes()])
kdtree_2021 = KDTree([graph_2021.nodes[node]['feature'] for node in graph_2021.nodes()])

distance_threshold = 200  # Define a distance threshold (in meters)
add_edges_with_kdtree(graph_1999, kdtree_1999, distance_threshold)
add_edges_with_kdtree(graph_2021, kdtree_2021, distance_threshold)
```

```python
from torch_geometric.data import Data
import torch

def convert_graph_to_torch(graph):
    edge_index = torch.tensor(list(graph.edges)).t().contiguous()
    x = torch.tensor([graph.nodes[node]['feature'] for node in graph.nodes()], dtype=torch.float)
    y = torch.tensor([graph.nodes[node]['label'] for node in graph.nodes()], dtype=torch.long)
    data = Data(x=x, edge_index=edge_index, y=y)
    return data

torch_data_1999 = convert_graph_to_torch(graph_1999)
torch_data_2021 = convert_graph_to_torch(graph_2021)
```

```python
print(torch_data_1999)
print(torch_data_2021)
```

```python
from torch_geometric.nn import GCNConv
import torch.nn.functional as F

class GCN(torch.nn.Module):
    def __init__(self, num_features, num_classes):
        super(GCN, self).__init__()
        self.conv1 = GCNConv(num_features, 16)
        self.conv2 = GCNConv(16, num_classes)

    def forward(self, data):
        x, edge_index = data.x, data.edge_index
        x = self.conv1(x, edge_index)
        x = F.relu(x)
        x = F.dropout(x, training=self.training)
        x = self.conv2(x, edge_index)
        return F.log_softmax(x, dim=1)
```

```python
from torch.optim import Adam

# Initialize the model and optimizer
model = GCN(torch_data_1999.num_features, len(le_1999.classes_))
optimizer = Adam(model.parameters(), lr=0.01)

# Training loop
for epoch in range(500):
    model.train()
    optimizer.zero_grad()
    out = model(torch_data_1999)
    loss = F.nll_loss(out, torch_data_1999.y)
    loss.backward()
    optimizer.step()
    print(f'Epoch: {epoch:03d}, Loss: {loss:.4f}') 
```

```python
from sklearn.metrics import accuracy_score
```

```python
# 1. Switch the model to evaluation mode
model.eval()

# 2. Make predictions on the validation set
with torch.no_grad():  # Temporarily set all the requires_grad flag to false
    output = model(torch_data_2021)  # Assuming torch_data_2021 is your validation set
    probabilities = F.softmax(output, dim=1)
    predicted_classes = probabilities.argmax(dim=1)

# 3. Calculate the loss and other performance metrics
loss = F.nll_loss(output, torch_data_2021.y)  # Negative Log Likelihood Loss
accuracy = accuracy_score(torch_data_2021.y.cpu(), predicted_classes.cpu())  # Assuming labels are on the CPU

print(f'Validation Loss: {loss.item()}')
print(f'Validation Accuracy: {accuracy}')
```
